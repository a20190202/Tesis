```python
import csv
import os
import subprocess
import urllib.parse

csv_file = '50modelos.csv'
download_script = 'bing_images_download.py'
#D:\Tesis\chromedriver-win64\chromedriver-win64
chromedriver_path = 'D:\\Tesis\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe'
#chromedriver_path = '/usr/bin/chromedriver'
parent_directory = 'imagenes'
num_images = 100

if not os.path.exists(parent_directory):
    os.makedirs(parent_directory)

with open(csv_file, newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        rank = row['Rank'].strip()
        name = row['Name'].strip()
        folder_name = f"{rank}_{name.replace(' ', '_')}"
        query = f"{name} 2023"
        query_encoded = urllib.parse.quote_plus(query)
        search_url = f"https://www.bing.com/images/search?q={query_encoded}&form=RESTAB&first=1"
        cmd = [
            'python', download_script,
            '-o', parent_directory,
            '-i', folder_name,
            '-u', search_url,
            '-l', str(num_images),
            '-cd', chromedriver_path
        ]
        print("Running command:", " ".join(cmd))
        subprocess.run(cmd)
```
