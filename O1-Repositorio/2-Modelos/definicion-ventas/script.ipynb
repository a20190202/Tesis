## Interest #1

```python
def get_average_interest(make, model):
    # Create the search term
    search_term = f"{make} {model}"
    
    # Get the Google Trends data
    pytrends.build_payload([search_term], timeframe='today 5-y', geo='PE')
    data = pytrends.interest_over_time()
    
    # Check if data is not empty and calculate average interest
    if not data.empty:
        return data[search_term].mean()  # Average interest over the last 5 years
    else:
        return None  # If no data is found

# Apply the function to create a new column in the DataFrame
df['average_interest'] = df.apply(lambda row: get_average_interest(row['make'], row['model']), axis=1)
```

## Chunks #3

```python
all_keywords = models[0:5]
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
for chunk in [ lst + [all_keywords[0]] for lst in list(chunks(all_keywords[1:], 2))]:
    print(chunk) 
```

```python
for chunk in [ lst + [all_keywords[0]] for lst in list(chunks(all_keywords[1:], 2))]:
    pytrends.build_payload(chunk, cat=0, timeframe='today 3-m', geo='PE')
    data = pytrends.interest_over_time()
    mean = data.mean()
    print(mean)
    print() 
```
