```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Cargar el archivo
archivo = "resultados.xlsx"  # Cambia por la ruta correcta
df = pd.read_excel(archivo)

# Seleccionar columnas de inters
modelos = df['ID']
metricas = ['mAP50', 'mAP50-95', 'Precision', 'Recall', 'F1']

# Configurar grfico
x = range(len(modelos))

for metrica in metricas:
    plt.figure(figsize=(8, 5))
    plt.bar(x, df[metrica], tick_label=modelos)
    plt.ylabel(metrica)
    plt.title(f'Comparacin de {metrica}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar archivo
archivo = "resultados.xlsx"  # Cambia por tu ruta
df = pd.read_excel(archivo)

# Columnas de inters
metricas = ['mAP50', 'mAP50-95', 'Precision', 'Recall', 'F1']

# Identificar el modelo YOLOv5 como lnea base
baseline_row = df[df['ID'].str.contains('YOLOv5', case=False)]
if baseline_row.empty:
    raise ValueError("No se encontr el modelo YOLOv5 en el archivo.")
baseline_values = baseline_row.iloc[0]

# Modelos sin incluir la lnea base para las barras
df_bar = df[~df['ID'].str.contains('YOLOv5', case=False)]

# Graficar cada mtrica
for metrica in metricas:
    plt.figure(figsize=(10, 6))
    
    x = range(len(df_bar))
    y = df_bar[metrica]

    plt.bar(x, y, tick_label=df_bar['ID'])

    # Agregar los valores encima de cada barra
    for i, v in enumerate(y):
        plt.text(i, v + 0.002, f"{v:.3f}", ha='center', va='bottom', fontsize=9)

    # Lnea base del modelo YOLOv5
    baseline_value = baseline_values[metrica]
    plt.axhline(baseline_value, color='red', linestyle='--', label=f'YOLOv5 (Lnea base): {baseline_value:.3f}')
    
    # Configurar lmites del eje Y
    ymin = min(y.min(), baseline_value)
    plt.ylim(bottom=max(ymin - 0.05, 0), top=1.0)

    plt.ylabel(metrica)
    plt.title(f'Comparacin de {metrica}')
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
